```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor as MLP
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
```

```python
X = np.genfromtxt('../datasets/2D_descriptors_X')
Y = np.genfromtxt('../datasets/2D_descriptors_Y')
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.05,random_state=1010)
print(X_train.shape,X_test.shape)
print(Y_train.shape,Y_test.shape)
```

```python
regr = MLP(solver='adam',activation='tanh',max_iter=500000,learning_rate_init=0.0000001,hidden_layer_sizes=(200,))
regr.fit(X_train,Y_train)
plt.figure(figsize = (5,5) )
plt.scatter(regr.predict(X_train),Y_train)
plt.plot(np.linspace(0,7),np.linspace(0,7))
plt.xlim((0,7))
plt.ylim((0,7))
plt.show()
```
