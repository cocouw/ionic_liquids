```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as MSE
```

```python
X = np.genfromtxt('../datasets/2D_descriptors_X')
Y = np.genfromtxt('../datasets/2D_descriptors_Y')
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.05,random_state=1010)
print(X_train.shape,X_test.shape)
print(Y_train.shape,Y_test.shape)
```

```python
a = np.logspace(-8,3, 10)
error_train = []
error_test = []
for regul in a:
    regr = linear_model.Lasso(alpha = regul)
    regr.fit(X_train,Y_train)
    error_train.append(MSE(regr.predict(X_train),Y_train))
    error_test.append(MSE(regr.predict(X_test),Y_test))
```

```python
plt.plot(np.linspace(-8,3,10),error_train)
plt.plot(np.linspace(-8,3,10),error_test)
plt.show()
```

```python
regr = linear_model.Lasso(alpha = 0)
regr.fit(X_train,Y_train)
plt.figure(figsize = (5,5) )
plt.scatter(regr.predict(X_train),Y_train)
plt.show()
```
